```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def streamplot2D(x_kiri,x_kanan,y_kiri,y_kanan,u,v,judul):
    x, y = np.meshgrid(np.linspace(x_kiri, x_kanan, 25),
                       np.linspace(y_kiri, y_kanan, 25))
    u = u(x,y)
    v = v(x,y)
    
    fig, ax = plt.subplots(figsize=(10, 6), dpi=100)
    plt.xlim(x_kiri, x_kanan)
    plt.ylim(y_kiri, y_kanan)
    plt.ylabel('y', rotation=0)
    
    ax.set_xlabel('$\mathbf{x}$').set_fontsize(15)
    ax.set_ylabel('$\mathbf{y}$').set_fontsize(15)
    
    ax.set_title('Stream Plot '+judul)
    plt.streamplot(x, y, u, v, color=np.linalg.norm(np.dstack([u, v]), axis=-1), cmap='autumn')
    plt.colorbar()
    return plt.show()
```

```python
x_kiri,x_kanan,y_kiri,y_kanan = -1,1,-1,1    #batas-batas nilai x dan y
u = lambda x,y: np.cos(x)                    #persamaan parametrik u(x,y)
v = lambda x,y: np.sin(y)                    #persamaan parametrik v(x,y) 
judul = 'Contoh 1'

streamplot2D(x_kiri,x_kanan,y_kiri,y_kanan,u,v,judul)
```

```python
x_kiri,x_kanan,y_kiri,y_kanan = -5,0,-3,3               #batas-batas nilai x dan y
u = lambda x,y: -4*(x+3)                                #persamaan parametrik u(x,y)
v = lambda x,y: -10*(y-2)*(y-1)*(y+2)*(3*y**2-2*y-4)    #persamaan parametrik v(x,y)
judul = "Contoh 2"

streamplot2D(x_kiri,x_kanan,y_kiri,y_kanan,u,v,judul)
```

```python
x_kiri,x_kanan,y_kiri,y_kanan = -0.35,0.35,-0.35,0.35    #batas-batas nilai x dan y
u = lambda x,y: -5*y-3*x                                 #persamaan parametrik u(x,y)
v = lambda x,y: 5*x-3*y                                  #persamaan parametrik v(x,y)
judul = "Contoh 3"

streamplot2D(x_kiri,x_kanan,y_kiri,y_kanan,u,v,judul)
```
