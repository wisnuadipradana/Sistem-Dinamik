```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib
import numpy as np
from numpy import sin,exp,cos,log,sqrt,pi
%matplotlib inline

def plot3D(x_kiri,x_kanan,y_kiri,y_kanan,z_kiri,z_kanan,offset_x,offset_y,offset_z,
           function,judul,jarak_colorbar,jarak_sumbuz,jarak_tickx,jarak_ticky,jarak_tickz,simpan): 
    x = np.arange(x_kiri, x_kanan, 0.2)
    y = np.arange(y_kiri, y_kanan, 0.2)
    xgrid, ygrid = np.meshgrid(x, y)
    function = function(xgrid,ygrid) 

    fig = plt.figure(figsize=(13,9), dpi=100)

    ax = fig.add_subplot(1, 1, 1, projection='3d')

    ax.view_init(15, 90+45)                     #view/angle dapat diubah ubah
    surf = ax.plot_surface(xgrid, ygrid, function, rstride=2, cstride=2, cmap=cm.coolwarm,
                           linewidth=1, antialiased=False, alpha=0.8)
    cset = ax.contour(xgrid, ygrid, function, zdir='x', 
                      offset=x_kanan+offset_x, cmap=matplotlib.cm.coolwarm)    
    cset = ax.contour(xgrid, ygrid, function, zdir='y', 
                      offset=y_kiri+offset_y, cmap=matplotlib.cm.coolwarm)   
    cset = ax.contour(xgrid, ygrid, function, zdir='z', 
                      offset=z_kiri+offset_z, cmap=matplotlib.cm.coolwarm)      

    plt.tight_layout()
    plt.xticks(np.arange(x_kiri, x_kanan, jarak_tickx))
    plt.yticks(np.arange(y_kiri, y_kanan, jarak_ticky))
    
    plt.rc('xtick', color='blue', labelsize='medium', direction='out')
    plt.rc('ytick', color='red', labelsize='medium', direction='in')
#     plt.rc('xtick.major', size=4, pad=20)
#     plt.rc('xtick.minor', size=2, pad=20)
#     plt.rcdefaults()    #uncomment jika ingin mengembalikan ke setinggan awal
    ax.tick_params(axis='z', which='major', pad=jarak_tickz)

    plt.subplots_adjust(top=0.8) 
    plt.title(judul, size=20, pad=-20)
    plt.rcParams['axes.titlepad'] = -20

    ax.set_xlabel('$x$').set_fontsize(20)
    ax.set_ylabel('$y$').set_fontsize(20)
    ax.set_zlabel('$z = $ $f(x,y)$', labelpad=jarak_sumbuz).set_fontsize(17)

    #Jika ingin memampatkan interval pada masing-masing sumbu x,y,z uncomment 3 kode dibawah
    #tetapi jika ingin tau ketinggian batas-batas z comment dahulu 3 kode dibawah
    ax.set_xlim3d(x_kiri-.5, x_kanan+.5)
    ax.set_ylim3d(y_kiri-.5, y_kanan+.5)
    ax.set_zlim3d(z_kiri, z_kanan)

    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

    fig.colorbar(surf, shrink=0.5, aspect=5, pad=jarak_colorbar)

    if simpan == True:
        ax.set_rasterized(True)
        fig.savefig('{}.png'.format(judul.translate(str.maketrans({' ':'', '.':''})).lower()), bbox_inches="tight", format='png', dpi=1000)
    return plt.show()
```

```python
x_kiri, x_kanan, y_kiri, y_kanan = -3,3,-3,3                 #batas kiri dan kanan interval x dan y
z_kiri, z_kanan = -.5, 1                                     #batas kiri dan kanan interval z
jarak_colorbar = 0.05                                        #jarak colorbar ke plot
jarak_sumbuz = 5                                             #jarak label z ke garis sumbu z
jarak_tickx,jarak_ticky,jarak_tickz = 1,1,2                  #jarak masing2 ticks ke garis sumbu x,y,z 
offset_x,offset_y,offset_z = .5,-.5,0                        #atur offset dari sumbu x, y dan z berturut-turut.
function = lambda x,y: 1/(1+np.abs(x-cos(4*pi*y))+np.abs(x**2+y**2-1))   #fungsi z=f(x,y)
judul = '3D Plot Contoh 1'
simpan = True                                                #simpan gambar True/False

plot3D(x_kiri,x_kanan,y_kiri,y_kanan,z_kiri,z_kanan,offset_x,offset_y,offset_z,
       function,judul,jarak_colorbar,jarak_sumbuz,jarak_tickx,jarak_ticky,jarak_tickz,simpan)
```
