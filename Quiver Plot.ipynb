# Quiver 2D

```python
import matplotlib.pyplot as plt
import numpy as np

def quiver2D(x_kiri,x_kanan,y_kiri,y_kanan,u,v,judul):
    X, Y = np.meshgrid(np.arange(x_kiri, x_kanan, .2),
                       np.arange(y_kiri, y_kanan, .2))
    U = u(X,Y)
    V = v(X,Y)

    fig, ax = plt.subplots(1, 3, figsize=(24,8), dpi=100, sharey=True)

    # Quiver kiri
    Q = ax[0].quiver(X, Y, U, V, units='width')
    qk = ax[0].quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                       coordinates='figure')
    
    # Quiver tengah
    Q = ax[1].quiver(X[::3, ::3], Y[::3, ::3], U[::3, ::3], V[::3, ::3],
                   pivot='mid', units='inches')
    qk = ax[1].quiverkey(Q, 0.9, 0.9, 1, r'$1 \frac{m}{s}$', labelpos='E',
                       coordinates='figure')
    ax[1].scatter(X[::3, ::3], Y[::3, ::3], color='r', s=5)    

    # Quiver kanan
    M = np.hypot(U, V)
    Q = ax[2].quiver(X, Y, U, V, M, units='x', pivot='tip', width=0.022,
                   scale=1 / 0.15)
    qk = ax[2].quiverkey(Q, 0.9, 0.9, 1, r'$1 \frac{m}{s}$', labelpos='E',
                       coordinates='figure')
    ax[2].scatter(X, Y, color='0.5', s=1)
    
    for i in range(3):
        ax[i].set_title(judul+" Bagian {}".format(i+1))
        ax[i].set_xlabel('$\mathbf{x}$', labelpad=5).set_fontsize(15)
    ax[0].set_ylabel('$\mathbf{y}$', labelpad=20, rotation=0).set_fontsize(15)

    plt.show()
```

```python
from numpy import *

x_kiri, x_kanan = 0, 2*pi     #batas-batas nilai x (x_kiri<=x<=x_kanan)
y_kiri, y_kanan = 0, 2*pi     #batas-batas nilai y (y_kiri<=y<=y_kanan)
u = lambda x,y: cos(x)        #persamaan parametrik u(x,y)
v = lambda x,y: sin(y)        #persamaan parametrik v(x,y)
judul = "Contoh 1 Quiver 2D"

quiver2D(x_kiri,x_kanan,y_kiri,y_kanan,u,v,judul)
```

```python
x_kiri, x_kanan = -5, 5      #batas-batas nilai x (x_kiri<=x<=x_kanan)
y_kiri, y_kanan = -4, 4      #batas-batas nilai y (y_kiri<=y<=y_kanan)
u = lambda x,y: 2*x-y**2     #persamaan parametrik u(x,y)
v = lambda x,y: -x-5*y       #persamaan parametrik v(x,y)
judul = "Contoh 2 Quiver 2D"

quiver2D(x_kiri,x_kanan,y_kiri,y_kanan,u,v,judul)
```

# Quiver 3D

```python
import matplotlib.pyplot as plt
import numpy as np

def quiver3D(x_kiri,x_kanan,y_kiri,y_kanan,z_kiri,z_kanan,u,v,w,judul):
    fig = plt.figure(figsize=(12,8), dpi=100)
    ax = fig.gca(projection='3d')
    ax.set_title(judul).set_fontsize(20)
    
    ax.set_xlabel('$\mathbf{x}$').set_fontsize(15)
    ax.set_ylabel('$\mathbf{y}$').set_fontsize(15)
    ax.set_zlabel('$\mathbf{z}$').set_fontsize(15)

    # Make the grid
    x, y, z = np.meshgrid(np.arange(x_kiri, x_kanan, 0.2),
                          np.arange(y_kiri, y_kanan, 0.2),
                          np.arange(z_kiri, z_kanan, 0.8))
    u = u(x,y,z)
    v = v(x,y,z)
    w = w(x,y,z)

    plt.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

    plt.show()
```

```python
from numpy import *

x_kiri, x_kanan, y_kiri, y_kanan = -0.8,1,-0.8,1                  #batas-batas nilai x dan y
z_kiri, z_kanan = -0.8, 1                                         #batas-batas nilai z
u = lambda x,y,z : sin(pi*x)*cos(pi*y)*cos(pi*z)                  #persamaan parametrik u(x,y,z)
v = lambda x,y,z : -cos(pi*x)*sin(pi*y)*cos(pi*z)                 #persamaan parametrik v(x,y,z)
w = lambda x,y,z : (sqrt(2.0/3.0)*cos(pi*x)*cos(pi*y)*sin(pi*z))  #persamaan parametrik w(x,y,z)
judul = 'Contoh 1 Quiver 3D'

quiver3D(x_kiri,x_kanan,y_kiri,y_kanan,z_kiri,z_kanan,u,v,w,judul)
```

```python
x_kiri, x_kanan, y_kiri, y_kanan = -2,2,-3,3             #batas-batas nilai x dan y
z_kiri, z_kanan = -0.1,1.6                               #batas-batas nilai z
u = lambda x,y,z: 4*(x+3)                                #persamaan parametrik u(x,y,z)
v = lambda x,y,z: 10*(y-2)*(y-1)*(y+2)*(3*y**2-2*y-4)    #persamaan parametrik v(x,y,z)
w = lambda x,y,z: -1                                     #persamaan parametrik w(x,y,z)
judul = 'Contoh 2 Quiver 3D'

quiver3D(x_kiri,x_kanan,y_kiri,y_kanan,z_kiri,z_kanan,u,v,w,judul)
```
