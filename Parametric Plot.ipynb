# Parametric Plot 2D

```python
%reset -f             
from sympy import *    
from sympy.plotting import plot_parametric
from sympy.plotting import plot3d_parametric_line
import math
import matplotlib.pyplot as plt

init_printing()        #for nice math output 

    #uncomment for plot in separate window
# %matplotlib qt       
    #uncomment for plot inline --- DEFAULT
%matplotlib inline     

t = symbols('t')

def parametric_plot(x,y,t_kiri,t_kanan,judul):
    plt.rcParams['figure.figsize'] = 12, 8
    plt.rcParams['font.size'] = 14
    plt.rcParams['legend.fontsize'] = 14
    plot_parametric(x, y, (t, t_kiri, t_kanan), line_color='blue', 
                    xlabel='$x(t)$',
                    ylabel='$y(t)$',
                    title='Parametric Plot '+judul)
    return plt.show()
```

```python
x = exp(-t)-12*t*exp(-3*t)             #fungsi x(t)
y = exp(-3*t)-6*t*exp(-3*t)            #fungsi y(t)
t_kiri, t_kanan = -pi, pi              #batas interval t dalam plot (t_kiri<=t<=t_kanan)             
judul = 'Contoh 1'

parametric_plot(x,y,t_kiri,t_kanan,judul)
```

```python
x = cos(t)                    #fungsi x(t)
y = sin(t)                    #fungsi y(t)
t_kiri, t_kanan = 0, 2*pi     #batas interval t dalam plot (t_kiri<=t<=t_kanan)
judul = 'Contoh 2'

parametric_plot(x,y,t_kiri,t_kanan,judul)
```

# Parametric Plot 3D

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

def plot3D_parametric(x,y,z,t,m_xoy,m_xoz,m_yoz,judul):
    plt.rcParams['legend.fontsize'] = 10
    plt.rcParams['font.size'] = 14

    fig = plt.figure(figsize=(10,8), dpi=100)
    ax = fig.gca(projection='3d')
    ax.set_title(judul) 

    ax.set_xlabel('$\mathbf{x}$').set_fontsize(15)
    ax.set_ylabel('$\mathbf{y}$').set_fontsize(15)
    ax.set_zlabel('$\mathbf{z}$').set_fontsize(15)
    
    ax.plot(x, y, 0*t+m_xoy, label='bidang $xoy$')        #atur pergesaran nilai m_xoy agar tepat di bidang xoy
    ax.plot(x, 0*t+m_xoz, z, label='bidang $xoz$')        #atur pergesaran nilai m_xoz agar tepat di bidang xoz
    ax.plot(0*t+m_yoz, y, z, label='bidang $yoz$')        #atur pergesaran nilai m_yoz agar tepat di bidang yoz
    ax.plot(x, y, z, color='black', label='parametric curve')
    ax.legend()

    return plt.show()
```

```python
t_kiri, t_kanan = 0.0001,3              #batas interval t dalam plot (t_kiri<=t<=t_kanan)
t = np.linspace(t_kiri, t_kanan, 100)   #100 adalah banyaknya step dari interval t_kiri ke t_kanan
x = t**3*np.log(np.exp(1)/t)            #x(t)
y = t**3                                #y(t)
z = t                                   #z(t)
m_xoy, m_xoz, m_yoz = 0,29,-2.7         #pergeseran nilai m
judul = 'Contoh 3D plot parametric 1'

plot3D_parametric(x,y,z,t,m_xoy,m_xoz,m_yoz,judul) 
```

```python
t_kiri, t_kanan = -10,20                #batas interval t dalam plot (t_kiri<=t<=t_kanan)
t = np.linspace(t_kiri, t_kanan, 100)   #100 adalah banyaknya step dari interval t_kiri ke t_kanan
x = t*np.cos(t)                         #x(t)
y = t*np.sin(t)                         #y(t)
z = t**2                                #z(t)

m_xoy, m_xoz, m_yoz = 0,19,-20         #pergeseran nilai m
judul = 'Contoh 3D plot parametric 2'

plot3D_parametric(x,y,z,t,m_xoy,m_xoz,m_yoz,judul) 
```
