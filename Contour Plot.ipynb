```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
%matplotlib inline  

def contourplot(x_kiri,x_kanan,y_kiri,y_kanan,function,judul,simpan):
    x = np.arange(x_kiri, x_kanan, 0.1)
    y = np.arange(y_kiri, y_kanan, 0.1)
    xgrid, ygrid = np.meshgrid(x, y)
    function = function(xgrid,ygrid)

    fig, ax = plt.subplots(1, figsize=(12,8), dpi=100, constrained_layout=True)
    CS = plt.contour(xgrid, ygrid, function, cmap=cm.coolwarm)
    plt.clabel(CS, inline=1, fontsize=10, fmt = '%.3f')
    plt.title(judul)
    
    if simpan == True:
        ax.set_rasterized(True)
        fig.savefig('{}.png'.format(judul), bbox_inches="tight", format='png', dpi=1000)
    return plt.show()
```

```python
x_kiri, x_kanan, y_kiri, y_kanan = -5,0,-10,10               #interval x dan y dalam plot kontur
function = lambda x,y: 2*(x+3)**2+5*((y-2)*(y+2)*(y-1))**2   #fungsi dari kontur plot
judul  = 'Contour Plot $2(x+3)^2+5((y-2)(y+2)(y-1))^2$'
simpan = False                                                # True/False untuk menyimpan gambar

contourplot(x_kiri,x_kanan,y_kiri,y_kanan,function,judul,simpan)
```

```python
x_kiri, x_kanan, y_kiri, y_kanan = -3,3,-3,3                                 #interval x dan y dalam plot kontur
function = lambda x,y: 1/(1+np.abs(x-np.cos(4*np.pi*y))+np.abs(x**2+y**2-1)) #fungsi dari kontur plot
judul  = 'Contoh Contour Plot 2'
simpan = True                                                              #True/False untuk menyimpan gambar

contourplot(x_kiri,x_kanan,y_kiri,y_kanan,function,judul,simpan)
```
